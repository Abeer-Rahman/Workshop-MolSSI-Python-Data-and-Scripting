# Numpy Lesson


```python
import numpy
help(numpy.genfromtxt)
```

```python
import os
import numpy
#build file path
distance_file=os.path.join('data','distance_data_headers.csv')
#Adv: No need to open and close file
distances=numpy.genfromtxt(distance_file,delimiter=",",dtype="unicode") #Arguments need not to be put in order
print(distances)
```

```python
type(distances)
```

```python
#We can slice a numpy array
headers=distances[0]
print(headers)
```

```python
data=distances[1:]
print (data)
```

```python
data
```

```python
data=data.astype(numpy.float)
```

```python
print(data[0,1])
```

```python
small_data=data[:10, :3]
print(small_data)
```

```python
#data_average=numpy.mean(data to take mean of)
thr4_atp=data[:,1]
print(thr4_atp)
avg_thr4_atp=numpy.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
#code for number of column in 2D array
num_columns=len(data[0,:])
print(num_columns)
##this is an alternative code for finding out no of columns:
#num_columns=print(len(headers))
```

```python
for i in range(num_columns):
    #print(i)
    column=data[:,i]
    col_headers=headers[i]
    avg_col=numpy.mean(column)
    print(F"{col_headers}: {avg_col:.3f}")   #number.3f------>gives 3 digits after decimal point,can only be done in F str print
```

# Plotting
This part will show how to create plot using the python library matplotlib

```python
import matplotlib.pyplot as plt
plt.figure()

plt.plot(data[:,1],label=headers[1])
plt.legend()
plt.xlabel("Simulation Frame")
plt.ylabel("Distance(Angstrom)")
#Saving the image file
plt.savefig("my_plot_hq.png",dpi=300)
```

```python
#Creating plot with multiple lines
import matplotlib.pyplot as plt
plt.figure()

plt.plot(data[:,1],label=headers[1])
plt.plot(data[:,2],label=headers[2])
plt.legend()
plt.xlabel("Simulation Frame")
plt.ylabel("Distance(Angstrom)")
#Saving the image file
plt.savefig("my_plot_hq_2_lines.png",dpi=300)
```

```python
#Creating plots with all columns of the data file
import matplotlib.pyplot as plt
plt.figure()
plt.xlabel("Simulation Frame")
plt.ylabel("Distances(Angstrom)")
for i in range(1,num_columns):
    plt.plot(data[:,i],label=headers[i]

plt.legend()
plt.savefig("my_plot_hq_all_col_lines.png",dpi=300)
```

```python
#Challenge: Making different plots for different columns and saving them in as different files
import matplotlib.pyplot as plt

for i in range(1,num_columns):
    plt.figure()
    plt.xlabel("Simulation Frame")
    plt.ylabel("Distances(Angstrom)")
    plt.ylim(0,20)
    plt.plot(data[:,i],label=headers[i])
    plt.legend()
    plt.savefig(F"{headers[i]}.png",dpi=200)


```
